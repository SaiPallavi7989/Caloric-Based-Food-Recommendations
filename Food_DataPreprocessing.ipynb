```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
#load data
df=pd.read_csv('Food Nutrition.csv')
#information
df.info()
#print column names
df.columns
```

```python
#total rows and columns
df.shape
```

```python
#first 5 rows
df.head()
```

```python
#Deleting Timestamp column
df=df.drop(columns=['Timestamp'])
```

```python
#Descriptive Statistics
df.describe(include='all')
```

```python
#convert into uppercase
df1= df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
print(df1.head())
```

```python
# Extract milliliter values into a new column 'SERVING_SIZE (in ml)'
df1['SERVING_SIZE (in ml)'] = df1['SERVING_SIZE (in g)'].str.extract(r'(\d+\s*ML)', expand=False)
df1.fillna(0,inplace=True)
print(df1.head(25))
```

```python
#convert ml values in SERVING_SIZE (in g) column to NaN
df1['SERVING_SIZE (in g)'] = pd.to_numeric(df1['SERVING_SIZE (in g)'], errors='coerce')
print(df1.head(25))
```

```python
#removing ml
df1['SERVING_SIZE (in ml)'] = df1['SERVING_SIZE (in ml)'].apply(lambda x: str(x).replace('ML', '') if isinstance(x, str) else x)
print(df1.head(25))
```

```python
df1['SERVING_SIZE (in ml)'] = pd.to_numeric(df1['SERVING_SIZE (in ml)'], errors='coerce')
print(df1.head(25))
```

```python
df1['SERVING_SIZE (in g)']=df1['SERVING_SIZE (in g)'].fillna(0)
print(df1.head(25))
```

```python
df1.to_csv('newfile.csv', index=False)
df=pd.read_csv('newfile.csv')
df.head()
```

```python
# List of columns to convert
columns_to_convert = [
    'CALORIC_VALUE (in g)',
    'PROTEIN (in g)',
    'MONOUNSATURATED_FAT (in g)',
    'POLYUNSATURATED_FAT (in g)',
    'CARBOHYDRATES (in g)',
    'VITAMIN A (in %)',
    'VITAMIN B1 (in %)',
    'VITAMIN B2 (in %)',
    'VITAMIN B3 (in %)',
    'VITAMIN B5 (in %)',
    'VITAMIN B6 (in %)',
    'VITAMIN B11 (in %)',
    'VITAMIN B12 (in %)',
    'VITAMIN E (in %)',
    'VITAMIN K (in %)',
    'WATER (in g)',
    'MAGNESIUM (in mg)',
     'MANGANESE (in mg)',
    'PHOSPHORUS (in mg)',
    'SELENIUM (in mg)',
    'ZINC (in mg)'
]

# Convert each column to numeric type
for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Use errors='coerce' to handle non-numeric values
```
