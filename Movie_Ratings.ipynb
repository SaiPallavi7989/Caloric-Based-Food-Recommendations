```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df = pd.read_csv('/content/movies.csv')
#dropping null values
df.dropna(inplace=True)
# Drop duplicate rows
df.drop_duplicates(inplace=True)
# Reset index after dropping rows
df.reset_index(drop=True, inplace=True)
df['ONE-LINE']=df['ONE-LINE'].str.replace('\n',' ')
df['STARS']=df['STARS'].str.replace('\n',' ')
df['GENRE']=df['GENRE'].str.replace('\n',' ')
df['RATING'] = df['RATING'].astype(float)
df['YEAR'] = df['YEAR'].str.replace(r'\D', '', regex=True)
df['primary_genre'] = df['GENRE'].apply(lambda x: x.split(',')[0])
```

```python
# Example: Histogram of ratings
plt.figure(figsize=(10, 6))
plt.hist(df['RATING'], bins=10, edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Example: Box plot of ratings by genre
plt.figure(figsize=(12, 8))
sns.boxplot(x='primary_genre', y='RATING', data=df)
plt.title('Box plot of Ratings by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Interactive scatter plot of rating vs. feedback sentiment
fig = px.scatter(df, x='RATING', y='Gross', color='GENRE', hover_name='MOVIES',
                 title='Rating vs. Gross Revenue by Genre')
fig.update_traces(marker_size=10)
fig.show()
```

```python
# Bar plot of average ratings by genre
plt.figure(figsize=(12, 8))
sns.barplot(x='primary_genre', y='RATING', data=df, ci=None)
plt.title('Average Ratings by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair plot for exploring relationships between numerical variables
sns.pairplot(df, vars=['RATING'], hue='GENRE',height=7)
plt.suptitle('Pair Plot of Ratings by Genre', y=1.02)
plt.show()
```

```python
# Histogram of ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['RATING'], bins=10, kde=True)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.show()
```
